```python
import os
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, Dataset
from torchvision import transforms, models
from PIL import Image
import matplotlib.pyplot as plt
from transformers import ViTForImageClassification, ViTImageProcessor

```

```python

# Image size for ViT & CNN
IMG_SIZE = 224

# Define Transformations
transform = transforms.Compose([
    transforms.Resize((IMG_SIZE, IMG_SIZE)),
    transforms.ToTensor(),
    transforms.Normalize([0.5], [0.5])
])

# Custom Dataset Class
class FakeRealDataset(Dataset):
    def __init__(self, root_dir, transform=None):
        self.root_dir = root_dir
        self.transform = transform
        self.image_paths = []
        self.labels = []

        for label, folder in enumerate(["real", "fake"]):
            folder_path = os.path.join(root_dir, folder)
            for img_name in os.listdir(folder_path):
                img_path = os.path.join(folder_path, img_name)
                self.image_paths.append(img_path)
                self.labels.append(label)

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        img_path = self.image_paths[idx]
        label = self.labels[idx]
        image = Image.open(img_path).convert("RGB")

        if self.transform:
            image = self.transform(image)

        return image, label

# Load Dataset
dataset_path = "train"  # Change to your dataset folder
dataset = FakeRealDataset(root_dir=dataset_path, transform=transform)
dataloader = DataLoader(dataset, batch_size=32, shuffle=True)

print(f"Loaded {len(dataset)} images successfully!")

```

```python
import torch.fft

def apply_fda(source_img, target_img, beta=0.01):
    # Convert images to frequency domain
    source_fft = torch.fft.fft2(source_img)
    target_fft = torch.fft.fft2(target_img)

    # Extract amplitude and phase
    source_amp, source_phase = torch.abs(source_fft), torch.angle(source_fft)
    target_amp, target_phase = torch.abs(target_fft), torch.angle(target_fft)

    # Blend source amplitude with target amplitude
    mixed_amp = (1 - beta) * source_amp + beta * target_amp

    # Reconstruct the adapted image
    adapted_fft = torch.polar(mixed_amp, source_phase)
    adapted_img = torch.fft.ifft2(adapted_fft).real

    return adapted_img

# Example FDA usage
sample_source, _ = dataset[0]  # Take one image
sample_target, _ = dataset[1]

fda_result = apply_fda(sample_source, sample_target)
plt.imshow(fda_result.permute(1, 2, 0))  # Convert tensor to image
plt.show()
```

```python
# Load ViT Model
vit_model = ViTForImageClassification.from_pretrained(
    "google/vit-base-patch16-224", 
    num_labels=2, 
    ignore_mismatched_sizes=True
)
# Move model to GPU if available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
vit_model.to(device)

# Define Loss and Optimizer
criterion = nn.CrossEntropyLoss()
optimizer = optim.AdamW(vit_model.parameters(), lr=2e-5)

print("ViT Model Ready for Training!")

```

```python
class CNNModel(nn.Module):
    def __init__(self):
        super(CNNModel, self).__init__()
        self.cnn = models.resnet18(pretrained=True)
        self.cnn.fc = nn.Linear(self.cnn.fc.in_features, 2)  # 2 classes

    def forward(self, x):
        return self.cnn(x)

# Load CNN Model
cnn_model = CNNModel().to(device)
optimizer_cnn = optim.AdamW(cnn_model.parameters(), lr=2e-4)

print("CNN Model Ready!")

```

```python
num_epochs = 5

for epoch in range(num_epochs):
    vit_model.train()
    cnn_model.train()
    total_loss_vit, total_loss_cnn = 0, 0

    for images, labels in dataloader:
        images, labels = images.to(device), labels.to(device)

        # FDA Image Adaptation (Batch-wise)
        if images.shape[0] > 1:  # Ensure at least 2 images in batch
            fda_images = []
            for i in range(images.shape[0] - 1):  # Apply FDA to pairs in batch
                fda_images.append(apply_fda(images[i], images[i+1]))

            fda_images.append(images[-1])  # Keep last image unchanged
            images = torch.stack(fda_images)  # Reconstruct batch

        # Debug: Check batch size consistency
        print("Image shape after FDA:", images.shape)
        print("Labels shape:", labels.shape)

        # ViT Training
        optimizer.zero_grad()
        vit_outputs = vit_model(images).logits  
        loss_vit = criterion(vit_outputs, labels)  # No shape mismatch now
        loss_vit.backward()
        optimizer.step()
        total_loss_vit += loss_vit.item()

        # CNN Training
        optimizer_cnn.zero_grad()
        cnn_outputs = cnn_model(images)
        loss_cnn = criterion(cnn_outputs, labels)
        loss_cnn.backward()
        optimizer_cnn.step()
        total_loss_cnn += loss_cnn.item()

    print(f"Epoch [{epoch+1}/{num_epochs}], ViT Loss: {total_loss_vit:.4f}, CNN Loss: {total_loss_cnn:.4f}")

print("Training Completed!")


```
